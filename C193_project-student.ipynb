```python
#Task 1 
#Import all the libraries and read cardiovascular.csv
import pandas as pd
import matplotlib.pyplot as plt

df= pd.read_csv('cardiovascular.csv')
df
```

# Average cholestrol found in various age groups

```python
#Task 2 
#Group by age and find the average cholesterol and make a dataframe out of it
grpbyage= df.groupby('age')['cholesterol'].mean().reset_index()
grpbyage
```

```python
#Task 3 
#Plot a line graph for various age group and their cholesterol 
label= grpbyage['age']
value= grpbyage['cholesterol']
fig= plt.subplots(figsize= (19,8))

plt.plot(label, value, label= "Average Colestrol", linewidth= 3.0)
plt.xlabel("Age")
plt.ylabel("Cholestrol")
plt.title("Average Cholestrol in various age group", fontsize= 20)

plt.legend()
plt.show()
```

# Correlation between systolic and diastolic blood pressure

```python
# Diastolic blood pressure - is the pressure in the arteries when the heart rests between beats
# Systolic blood pressure - the pressure in your arteries when your heart beats

#predefine code for image
from IPython.core.display import Image
Image(filename='blood pressure readings chart.jpg') 
#predefine code end
```

```python
#Task 4
#Group by age and find maximum systolic blood pressure and create a dataframe project out of it

grpbyagesystolic= df.groupby('age')['systolic_blood_pressure'].max().reset_index()
grpbyagesystolic.dtypes
```

```python
#Task 5
#Group by age and find maximum diastolic blood pressure and create a dataframe project out of it

grpbyagediastolic= df.groupby('age')['diastolic_blood_pressure'].max().reset_index()
grpbyagediastolic
```

```python
#Task 6
#Plot a line graph to show a Correlation between systolic and diastolic blood pressure
label1= grpbyagesystolic['age']
value1= grpbyagesystolic['systolic_blood_pressure']

label2= grpbyagediastolic['age']
value2= grpbyagediastolic['diastolic_blood_pressure']
fig= plt.subplots(figsize= (19,8))

plt.plot(label1, value1, label= "Average systolic_blood_pressure", linewidth= 3.0)
plt.plot(label2, value2, label= "Average diastolic_blood_pressure", linewidth= 3.0)
plt.xlabel("Age")
plt.ylabel("Blood Pressure")
plt.title("Correlation between systolic and diastolic blood pressure", fontsize= 20)

plt.legend()
plt.show()
```
